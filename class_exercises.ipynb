## Question 2 : T2 vs T2* imaging 

You are reviewing the brain images of a subject who has experienced a mild traumatic brain injury (mTBI). Upon comparing the T2 and T2* images for this subject, you notice that one of the images contains hypointensities (dark regions) while the other does not. 

Which of the following statements best describes the image with hypointensities? 

**Answer A** : 
The image with hypointensities is the T2 image, which is more sensitive to detecting lesions associated with inflammation and edema. 

**Answer B** :
The image with hypointensities is the T2 image, where the hypointensities are likely artifacts caused by magnetic field inhomogeneities. 

**Answer C** :
The image with hypointensities is the T2 image, which can highlight lesions from hemorrhages and calcifications.

## Question 3 : Quantitative susceptibility Mapping (QSM)

You then decide to use Quantitative susceptibility mapping (QSM) to see if you are able to see differences between healthy controls and mTBI like T2*w.

Like we have seen in chapter 2, QSM allows us to quantify local tissue magnetic susceptibility. A positive susceptibility value indicates the presence of paramagnetic substances, while negative susceptibility values suggests a diamagnetic substance. 

Given this information, if the hypointensities from the T2*w image correspond to a microhemorrhage and a calcification, how should we expect them to show on the QSM image?

**Answer A** : The microhemorrhage and the calcification would both appear hypointense. 

**Answer B** : The microhemorrhage and the calcification would both appear hyperintense.  

**Answer C** : The microhemorrhage would appear hyperintense while the calcification would appear hypointense.

**Answer D** : The microhemorrhage would appear hypointense and the calcification, hyperintense.
